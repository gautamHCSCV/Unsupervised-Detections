```python
import numpy as np
import os
import cv2
import shutil
```

```python
def extract_frames(video_path, output_folder):
    # Open the video file
    video = cv2.VideoCapture(video_path)

    os.makedirs(output_folder, exist_ok=True)
    frame_count = len(os.listdir(output_folder))+1

    # Read the video frames
    while True:
        for i in range(7):
            ret, frame = video.read()

        if not ret:
            break

        # Save the frame as an image file
        frame_path = os.path.join(output_folder, f"frame{frame_count:04d}.jpg")
        cv2.imwrite(frame_path, frame)
        frame_count += 1

    # Release the video object
    video.release()

    print(f"Frames extracted: {frame_count}")
```

```python
# path = '/content/drive/MyDrive/New Product/Mind-Vision/'
# for i in range(2,len(os.listdir(path))):
video_path = '/content/drive/MyDrive/Loading_Unloading/yolo_model/Videos/val4.mp4'
output_folder = 'Frames2'

# Extract frames from the video
extract_frames(video_path, output_folder)
```
